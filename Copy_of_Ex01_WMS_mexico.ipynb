```python
# Import the pandas library for data manipulation and analysis
import pandas as pd
```

```python
# Import 'files' from Google Colab to enable downloading files from the Colab environment
from google.colab import files
```

```python
# Define the URL of the CSV file stored on Dropbox (set to direct download)
myfile='https://www.dropbox.com/scl/fi/zoanfqe59tuwl3g7r59zo/wms_da_textbook.csv?rlkey=kpz5o6urg14z974lhh7ic5c33&st=1xlxhsd5&dl=1'
```

```python
# Read the CSV file from the URL into a pandas DataFrame
df = pd.read_csv(myfile)
```

```python
# Display the first few rows of the dataset to verify that it loaded correctly
print(df.head())
```

```python
# Display the last few rows of the dataset
print(df.tail())
```

```python
# Filter the rows
# Keep only rows that meet all the following conditions:
# 1. The 'country' column equals 'Mexico'
# 2. The 'wave' column equals 2013
# 3. The 'emp_firm' (number of employees in the firm) is between 100 and 5000, inclusive
df_mexico = df[(df['country'] == 'Mexico') & (df['wave'] == 2013) & (df['emp_firm'] >= 100) & (df['emp_firm'] <= 5000)]
```

```python
# Print the total number of observations in the dataset
print("Number of observations:", len(df))
```

```python
# Print the total number of observations in the filtered dataset
print("Number of observations:", len(df_mexico))
```

```python
# Save the filtered dataset to a new CSV file
df_mexico.to_csv('wms_mexico_2013.csv')
```

```python
# Download the saved CSV file to your local machine (works in Google Colab)
files.download('wms_mexico_2013.csv')
```
